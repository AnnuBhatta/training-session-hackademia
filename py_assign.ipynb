```python
#TASK 5 task5_for_loops

n = int(input("Enter number of subjects: "))

total = 0
marks_list = []

for i in range(n):
    marks = float(input(f"Enter marks for subject {i + 1}: "))
    marks_list.append(marks)
    total += marks
average = total / n
highest = max(marks_list)
lowest = min(marks_list)

print("Total marks:", total)
print("Average marks:", average)
print("Highest marks:", highest)
print("Lowest marks:", lowest)
```

```python
#TASK 6 task6_while_validation

marks = float(input("Enter marks (0-100): "))
while marks < 0 or marks > 100:
    print("Invalid input! Marks must be between 0 and 100.")
    marks = float(input("Enter marks again (0-100): "))

print("Valid marks entered:", marks)
```

1. Why is validation better handled with while loops than for loops?
while loops repeat until the input is correct.
for loops repeat a fixed number of times, so they may stop even if the input is wrong.
This makes while better for input validation.

```python
#TASK 7 task7_strings

sentence = input("Enter a sentence: ")

vowels = 0
consonants = 0
digits = 0
spaces = 0

cleaned_sentence = " ".join(sentence.split()) 
lower_sentence = cleaned_sentence.lower()

for char in lower_sentence:
    if char in "aeiou":
        vowels += 1
    elif char.isalpha():
        consonants += 1
    elif char.isdigit():
        digits += 1
    elif char == " ":
        spaces += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Spaces:", spaces)
print("Cleaned Sentence:", cleaned_sentence)
print("Uppercase:", cleaned_sentence.upper())
print("Lowercase:", cleaned_sentence.lower())
```

```python
#TASK 8 task8_functions.py

def calculate_average(marks_list):
    return sum(marks_list) / len(marks_list)

def determine_grade(avg):
    if avg >= 80:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 40:
        return "C"
    else:
        return "Fail"

def display_result(avg, grade):
    print(f"Average Marks: {avg:.2f}")
    print(f"Grade: {grade}")

n = int(input("Enter number of subjects: "))
marks = []
for i in range(n):
    marks.append(float(input(f"Enter marks for subject {i + 1}: ")))

average = calculate_average(marks)
grade = determine_grade(average)
display_result(average, grade)
```

1. Why is returning values better than printing inside functions?
Returning values lets the main program use the results in different ways (like calculations, decisions, or printing).
Printing inside the function only shows the result and can’t be reused later.

```python
#TASK 9 task9_menu_system

def enter_student_details():
    global student_name, student_age
    student_name = input("Enter student name: ")
    student_age = int(input("Enter student age: "))
    print("Student details saved!\n")

def enter_marks():
    global marks_list
    n = int(input("Enter number of subjects: "))
    marks_list = []
    for i in range(n):
        mark = float(input(f"Enter marks for subject {i + 1}: "))
        while mark < 0 or mark > 100:  # input validation
            print("Invalid marks! Must be 0-100.")
            mark = float(input(f"Enter marks for subject {i + 1}: "))
        marks_list.append(mark)
    print("Marks saved!\n")

def calculate_average(marks):
    return sum(marks) / len(marks)

def determine_grade(avg):
    if avg >= 80:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 40:
        return "C"
    else:
        return "Fail"

def view_summary():
    if not student_name or not marks_list:
        print("No student details or marks found!\n")
        return
    avg = calculate_average(marks_list)
    grade = determine_grade(avg)
    print(f"\n--- Result Summary ---")
    print(f"Name: {student_name}")
    print(f"Age: {student_age}")
    print(f"Marks: {marks_list}")
    print(f"Average: {avg:.2f}")
    print(f"Grade: {grade}\n")

student_name = ""
student_age = 0
marks_list = []

while True:
    print("----- Student Utility Menu -----")
    print("1. Enter student details")
    print("2. Enter marks")
    print("3. View result summary")
    print("4. Exit")
    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        enter_student_details()
    elif choice == "2":
        enter_marks()
    elif choice == "3":
        view_summary()
    elif choice == "4":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice! Please enter 1-4.\n")
```
